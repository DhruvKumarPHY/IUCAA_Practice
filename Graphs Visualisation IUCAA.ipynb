```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,6.28,0.001)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.xlabel("Theta(angle in Radians)")
plt.ylabel("sin(Theta)")
plt.title("sin plot with precession of 0.001")           
```

```python
plt.plot(x,y)
xlab = plt.xlabel(r"$\theta$(Angle in Radians)")
ylab = plt.ylabel(r"sin($\theta$)")
thetitle = plt.title("More Colourful and Detailed");thetitle.set_fontweight('bold')

ylab.set_color('blue'); xlab.set_color('purple'); xlab.set_style('italic'); ylab.set_style('italic')
```

```python
another_x = x+1
plt.plot(another_x,another_x**4)
plt.title("X**4 vs X")
```

```python
#let us make it bit better 

plt.plot(another_x,another_x**4)
plt.title(r"X$^4$ verses X")
plt.xlim(min(another_x), max(another_x))   
plt.yscale('log')  
plt.show()  
```

```python
plt.rcParams['figure.figsize'] = (5,3.5)
plt.xkcd()
plt.plot(x,y)
plt.plot(x,np.cos(x))
xlab = plt.xlabel(r"$\theta$(Angle in radians)")
plt.rcdefaults()
plt.rcParams['figure.figsize'] = (5,3.5)
```

```python
plt.plot(x,y)
x_spaced= np.arange(0,2.0*np.pi,0.2)
plt.plot(x_spaced, np.sin(x_spaced), marker= 'o', color='red',linestyle='none')
plt.plot(x_spaced, np.cos(x_spaced), 'b*')
```

```python
plt.plot(x,y)
x_spaced= np.arange(0,2.0*np.pi,0.2)
plt.plot(x_spaced, np.sin(x_spaced), 'gs', ls='-.', lw='1', alpha=0.5)
```

```python
plt.plot(x,y ,label='Sin(x)')
plt.plot(x, np.cos(x))
y_scatter = y + np.random.normal(0.0,0.05,len(x))
plt.plot(x, y_scatter, 'b,', ls='none', label='Fake data with scatter')
plt.legend()
```

```python
y_error = np.random.normal(0.0,0.2,len(x_spaced))
x_error = np.repeat(0.1,len(x_spaced))
plt.errorbar(x_spaced, np.sin(x_spaced), yerr = y_error, xerr= x_error, marker='x', ecolor='r')
```
