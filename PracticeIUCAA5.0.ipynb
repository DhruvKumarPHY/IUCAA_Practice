```python
#scatter plot

import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('Solarize_Light2')
dist1 =np.random.normal(0.,1.,200)
dist2 =np.random.normal(0.,1.,200)
plt.scatter(dist1,dist2,color=  'r', marker='*', alpha=0.9)
plt.style.use('default')
```

```python
dist1 = np.random.normal(6.,2.,10000000)
n,bins,patches = plt.hist(dist1,bins= 50,color='y', density=False, histtype = 'step')
plt.grid()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.rcParams['figure.figsize'] = (3,3)
plt.subplot(221); plt.plot(t1, f(t1),'rs',t2,f(t2),'b')
plt.subplot(222); plt.plot(t2,np.cos(2*np.pi*t2), 'r-')
plt.subplot(223); plt.plot(t2,np.sin(2*np.pi*t2), 'y:')
plt.subplot(224); plt.plot(t2,t2**2,'g,')
```

```python
# Contour Graph 

def f(x,y):
    return(1-x/2 + x**5 + y **3) * np.exp(-x**2 - y**2)
n= 256 ; x= np.linspace(-3,3,n); y= np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)
plt.axes([0.025,0.025,0.95,0.95])
plt.contourf(X,Y , f(X,Y), 8, alpha= 0.8, cmap=plt.cm.hot)
C = plt.contour(X,Y , f(X,Y) ,8, color= 'black',linewidths=0.5 )
plt.clabel(C, inline= 1,fontsize =20)
```

```python
# 3D Plotting

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(5,3.5))
ax = fig.add_subplot(111,projection = '3d')
r = np.linspace(0,1.25,70)
p = np.linspace(0,2*np.pi,70)
R,P = np.meshgrid(r,p)
X,Y = R * np.cos(P), R * np.sin(P)
Z = ((R**2 -1)**2)
ax.plot_surface(X,Y,Z, rstride=1, cstride=1, cmap=plt.cm.hot)
```
